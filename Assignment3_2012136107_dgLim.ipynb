### 문제 2

- 문자열 S = 'Hello World and Python'에 대해 다음 요구사항에 대한 Python 코드를 제시하시오.
  - 1) 단어의 순서가 역순으로 되어진 문자열 ('Python and World Hello')을 만들고
  - 2) 1)의 결과 문자열에 대해 메소드 split과 join을 이용하여 공백을 모두 없엔 문자열 'PythonandWorldHello'을 만드시오.

```python
S = 'Hello World and Python'

str1 = S.split()     # 리스트str1에 문자열 S를 공백을 기준으로 쪼갠 단어들을 원소들로 갖는다.
str1 = str1[::-1]    # 리스트str1의 원소들을 역순으로 대입
S = " ".join(str1)   # " "즉, 공백을 띄우고 str1의 원소들을 합쳐 문자열을 만든 S

print(S)

print()

str2 = S.split()     # 문자열 S을 다시 공백을 기준으로 쪼개 리스트str2에 대입
S = "".join(str2)    # "" 즉, 공백 없이 str2의 원소들을 합쳐 문자열 S를 만듦

print(S)
```

#### [설명]

문자열 메소드 split()는 문자열을 공백을 기준으로 쪼갠 원소들을 가지는 리스트를 만드는 것이고, 문자열 메소드 join()은 입력받은 리스트의 원소들을 합치는 메소드이다. 이 때, " ".join(입력 리스트)는 각 원소 사이에 공백을 가지고 문자열로 합치는 것

### 문제 4

- 다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)


      s = '  first star   :   second star   :    third star  '
      l = ['first star', 'second star', 'third star']


  - 1) for ~ in 문을 이용한 코드를 작성하시오.
  - 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '
str1 = s.split(':')      # 문자열을 ':'를 기준으로 쪼개 str1 리스트의 원소로 갖는다.  

                         # 1) for ~ in문을 이용한 코드
for x in range(len(str1)):
    l[x] = str1[x].strip()

print(l)
```

#### [설명]

 우선, 문자열을 :을 기준으로 쪼개 원소로 갖는 리스트 str1을 만든다.
 
 l은 최종적으로 ':', 공백이 제거된 원소들을 갖는 리스트이므로, 선언과 동시에 길이를 정해준다.
 
 for~in문은 범위(range(len(str1)), 즉 3)이며, 
 리스트 l의 각 인덱스에 맞게 ':'을 기준으로 쪼갠 문자열 str1의 원소들에서 양 공백을 제거한(strip()) 원소를 갖는다.

```python
# 2) 리스트 내포(list comprehension)을 이용한 코드

s = '  first star   :   second star   :    third star  '
str1 = s.split(':')      # 문자열을 ':'를 기준으로 쪼개 str1 리스트의 원소로 갖는다.  

l = [str1[x].strip() for x in range(len(str1))]

print(l)
```

#### [설명]

리스트 내포(list comprehension)를 이용한 코드는

[표현식 for ~ in]으로 표현할 수 있다.

따라서 l = [str1[x].strip() for x in range(len(str1))]



### 문제 5

- 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

  - 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
  
        >>> addall([1])
        1
        >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
        45

```python
# 1) for-in
def addall(list_1):
    sum = 0
    for x in range(len(list_1)):
        sum = sum + list_1[x]
    return print(sum)

addall([1])
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 1) 리스트 내포
def addall(list_1):
    return print(sum([list_1[x] for x in range(len(list_1))]))

addall([1])
addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

  - 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
  
        >>> addallodd([1])
        1
        >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
        25

```python
# 1) for-in
def addallodd(list_1):
    sum = 0
    for x in range(len(list_1)):
        if(list_1[x]%2 == 1):
            sum = sum + list_1[x]
    return print(sum)

addallodd([1])
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 1) 리스트 내포
def addall(list_1):
    return print(sum([list_1[x] for x in range(len(list_1)) if list_1[x]%2 == 1]))

addallodd([1])
addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

#### [설명]

리스트 변수를 인자로 받는 함수 addall, addallodd를 만들 때, 함수문 선언을 하고, 그 내부에 for~in, 리스트 내포 두 가지 방법으로 제시하는 문제이다.
 우선, for~in 방법으로 문제를 해결할 때는 내장함수 sum()을 사용하지 않고, 리스트의 각 원소의 값의 합을 저장할 sum = 0을 선언하여 해결한다.
 sum = sum + list[x]
  addallodd 함수를 for~in 방법으로 해결 할 때는 조건만 하나 달아주면 된다. 홀수 이면 더하도록 if(list[x]%2 == 1)일 때만, 더하면 된다.
 둘째로, 리스트 내포 방식으로 문제를 해결하기 위해서는 내장함수 sum()을 이용해야 하는데, 리스트 내포 방식의 출력은 리스트이므로 sum()을 이용해서 리스트 결과값의 각 원소의 합을 출력하도록 한다.
 리스트 내포에서 조건문을 쓰는 방법은 for~in과 달리 위 문제와 같이 쓰면 된다.

  - 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.

    - 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] -->['world', 'hello', 'python', 'world']
    
    - 모든 punctuation을 확인하는 방법은 아래와 같다.

        import string
        
        print(string.punctuation)


  - 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
  
    - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']


  - 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.

    - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}

```python
import string

from urllib import request 
 
url = 'http://www.naver.com'

f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')

source.strip()                                  #양 쪽 끝에 있는 공백을 없앤다.
a = 0
b = 0
count = 0                                       #HTML 구문, 코드를 제외한 문자의 개수
t1 = 0
t2 = 0
c = []

i = 0
# javaScript 태그 없애기
while(i != 2):
    while(t1 != -1):                                 # a != -1의 조건 기준은 source.find('<', b)에서 b 인덱스 이후로 < 문자가 더이상 존재하지 않을 때, a에다가 -1을 리턴해준다.
        t1 = source.find('<script', t2)
        t2 = source.find('</script>', t1)

        if(t1 != t2):
            source = source[:t1] + source[t2:]

    t1 = 0
    t2 = 0
    i = i + 1
    
i = 0
# CSS 태그 없애기
while(i != 2):
    while(t1 != -1):                                 # a != -1의 조건 기준은 source.find('<', b)에서 b 인덱스 이후로 < 문자가 더이상 존재하지 않을 때, a에다가 -1을 리턴해준다.
        t1 = source.find('<style', t2)
        t2 = source.find('</style>', t1)

        if(t1 != t2):
            source = source[:t1] + source[t2:]

    t1 = 0
    t2 = 0
    i = i + 1

while(a != -1):                                 # a != -1의 조건 기준은 source.find('<', b)에서 b 인덱스 이후로 
                                                # < 문자가 더이상 존재하지 않을 때, a에다가 -1을 리턴해준다.
    a = source.find('<', b) 
    b = source.find('>', a) 
    a1 = source.find('<', b)

    if((a1 > b) and (a1 != a) ):                # > 이 사이에 있는 것이 문자열이라는 조건 <
        count = count + len(source[(b+1):a1].split())        # 그 사이에 있는 것이 문자열인데 그 안에 단어가 몇개 있는지를 공백 기준 쪼갠다.
        if(len(source[(b+1):a1].split()) > 0):
            c = c + source[(b+1):a1].split()
            
c = ' '.join(c)        # 리스트는 구두문자를 제거하기 위한 translate를 사용할 수 없으므로 문자열로 만든다. 이 때, 각 원소의
                       # 구분은 공백문자로 한다. 이후에 공백문자를 기준으로 다시 리스트를 만들 것이다.
translator = str.maketrans('', '', string.punctuation)
c = c.translate(translator)
c = c.split()
k = 0
result = []

#리스트 내에 구두문자를 제거한 단어들을 삽입한다.
while(k < len(c)):
    result.append((c[k],c.count(c[k])))
    k = k + 1

    
#중복 문자 삭제
result = set(result)

#리스트로의 형변환
result = list(result)

# 사전화
k = 0
list_key = []
list_value = []

while(k < len(result)):
    list_key.append(result[k][0])
    list_value.append(result[k][1])
    k = k + 1
 
#key, value(반복횟수)를 리스트로 각각 저장했었기 때문에, 리스트와 리스트 간
#mapping을 위해 zip()사용.

t = zip(list_key, list_value)

#zip 후, 사전으로 형변환
t = dict(t)

print(count) #중복을 제외한 단어의 개수
print()
print(t)
```

-  우선, 전체적인 url에서 '<', '>'을 구분하여 단어를 추리기 전에, java Script와 CSS로 이루어진 태그를 먼저 처리하고 나서 해야 원하는 결과 값을 얻을 수 있다고 판단했다. 
 
-  따라서, < script>가 시작하는 인덱스에서 < /script>로 끝나는 java Script의 태그와 그 안에 있는 내용을 먼저 제거했다. 제거 방법은 슬라이싱을 이용했는데,
 t1은 자바 스크립트가 시작하는 인덱스이며, t2는 자바 스크립트가 끝나는 인덱스로 그 사이의 내용을 없애기 위해 [:t1] + [t2:]의 범위를 취했다.
 
-  마찬가지로, CSS 태그인 < style, < /style>또한 마찬가지로 제거했다.
 
-  그렇게 나온 source를 가지고 지난 시간에 했던 html 태그를 제외한 문자들을 구하는 내용을 넣었다.

-  이렇게 나온 단어들에서 구두문자를 제거해야한다.

-  translate(None, string.punctuation)은 구두점이 스트링에 포함되어 있으면 모두 제거하라는 의미이나, 이 방법은 python 2.* version에서 사용하던 방법이었고, 
 python3에서는 str.maketrans('', '', string.punctuation), translate(c)를 이용하는 방법으로 사용해야만 했다.

- 이 때, 리스트를 받아오지 못하므로 문자열로 전환하기 위해 join을 이용했다.
 
- 그렇게 구두문자를 제거한 c는, 사전에 넣고 key로 단어, value로 그 단어의 중복 횟수를 넣을 것이지만, 바로 사전에 입력하면 중복된 내용이 기입되므로, 순서상 그 단어의 중복횟수를 저장하고 나서 set을 이용해 중복을 제거해야한다. 따라서, key, value에 해당하는 리스트를 저장하고, 중복을 제거하고 나서 따로따로 key, value를 만들어 사전으로 합치도록 zip(), dict()을 이용하였다.
